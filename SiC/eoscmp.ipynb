# Comparison of EOS data and models

- Strength probably significant on Hugoniot below 60 GPa.
- LEOS 2130 too stiff above 100 GPa: phase change to NaCl.
- CASTEP reasonable for cZnS and NaCl structures (LDA a little overbound). Predicts B3->B1 ~69 GPa.
- Need to add ambient isotherm comparison.

```python
import matplotlib.pyplot as plt
import numpy as np
from plotutils import *
```

## Principal Hugoniot

```python
plt.xlabel('density (g/cm^3)'); plt.ylabel('pressure (GPa)')

# alt graphs
#plt.yscale('log'); plt.ylim(bottom=0.1,top=1e8)
#plt.yscale('log'); plt.xlim(left=7.0,right=20.0); plt.ylim(bottom=1.0,top=1e3)
plt.xlim(left=3.0,right=5.0); plt.ylim(bottom=0.0,top=2e2)

# experimental data
with open('hugoniot/marsh.dat') as fp: marsh=readHugrho0usup(fp)
plt.scatter(marsh[0]['rho'],marsh[0]['p'],label='Marsh')
with open('hugoniot/vanthiel.dat') as fp: vanthiel=readHugrho0usup(fp)
plt.scatter(vanthiel[0]['rho'],vanthiel[0]['p'],label='van Thiel')
#plotfilepts('hugoniot/trunin.dat',ix=3,iy=4,label='Trunin')

# models
plotfileline('models/leos2130/hugoniot.dat',ix=0,iy=5,label='LEOS 2130')

plt.legend()
plt.show()
```

## Principal isentrope

```python
plt.xlabel('density (g/cm^3)'); plt.ylabel('pressure (GPa)')
plt.yscale('log')

# alt graphs
#plt.ylim(bottom=0.1,top=1e4)
plt.xlim(left=3.0,right=12.0); plt.ylim(bottom=10.0,top=2e3)

# experimental data
plotfileerrorbars('ramp/Kim2022_supptab2.dat',ix=5,iy=3,idx=6,idy=4,label='ramp: Kim 2022 pi')

# models
plotfileline('models/leos2130/isentrope.dat',ix=0,iy=2,label='LEOS 2130')

plt.legend()
plt.show()
```
