# Comparison of EOS data and models

- LEOS 830 is a reasonable match to Hugoniot data above ~40 GPa. The cold curve is consistent with DFT predictions of the bcc structure above ~100 GPa.
- The multiphase EOS models by Jeff Colvin and Marina Bastea are a much better match to lower-pressure shock data, but are limited in range.
- The Colvin model extends to higher density than the Bastea model, but is implausibly soft above ~10 GPa.
- The cold curve in the Bastea model has anomalous behavior in mixed-phase regions.
- Need to add DAC data and ambient isotherm comparison.
