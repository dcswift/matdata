# Comparison of EOS data and models

- SESAME 2160 is softer than other EOS models above a few tens of GPa.
- LEOS 500, 501, 502 don't capture softer p-rho behavior caused by phase transitions 10<p<80 GPa.
- LEOS 503 seems to have anomalously high pressures in the mixed-phase ~15 GPa on the cold curve.
- Need to add DAC data, ambient isotherm comparison.

```python
import matplotlib.pyplot as plt
import numpy as np
from plotutils import *
```

## Principal Hugoniot

```python
plt.xlabel('density (g/cm^3)'); plt.ylabel('pressure (GPa)')

# alt graphs
#plt.yscale('log'); plt.ylim(bottom=0.1,top=1e8)
#plt.yscale('log'); plt.xlim(left=7.0,right=20.0); plt.ylim(bottom=1.0,top=1e3)
plt.xlim(left=7.0,right=12.0); plt.ylim(bottom=0.0,top=1e2)

# experimental data
with open('hugoniot/marsh.dat') as fp: marsh=np.transpose(readMarsh(fp))
plt.scatter(marsh[0],marsh[1],label='Marsh')
with open('hugoniot/vanthiel.dat') as fp: vanthiel=np.transpose(readcompauto(fp,pscale=0.1)) # kbar->GPa
plt.scatter(vanthiel[0],vanthiel[1],label='van Thiel')
#plotfilepts('hugoniot/trunin.dat',ix=3,iy=4,label='Trunin')

# models
plotfileline('models/ses2160/hugoniot.dat',ix=0,iy=5,label='SESAME 2160')
plotfileline('models/ses2161/hugoniot.dat',ix=0,iy=5,label='SESAME 2161')
plotfileline('models/ses2162v1/hugoniot.dat',ix=0,iy=5,label='SESAME 2162')
plotfileline('models/leos500/hugoniot.dat',ix=0,iy=5,label='LEOS 500')
plotfileline('models/leos501/hugoniot.dat',ix=0,iy=5,label='LEOS 501')
plotfileline('models/leos502/hugoniot.dat',ix=0,iy=5,label='LEOS 502')
plotfileline('models/leos503/hugoniot.dat',ix=0,iy=5,label='LEOS 503')

#plotfileline('models/inferno81k/hugoniot.dat',ix=0,iy=5,label='Inferno 81k')

plt.legend()
plt.show()
```
